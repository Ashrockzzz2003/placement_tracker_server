```python
import pandas as pd
```

```python
df = pd.read_csv('2022.csv')
df
```

```python
df['studentEmail'] = df['studentRollNo'].str.lower() + '@cb.students.amrita.edu'
df['studentPassword'] = df['studentEmail'].str.lower() + "_" + df['studentSection']
df
```

```python
# studentRollNo,studentEmail,studentPassword,studentName,studentSection,studentBatch,studentDept,isHigherStudies,isPlaced

f = open("insertStudentData.sql", "w")

for index, row in df.iterrows():
    f.write(f"INSERT INTO studentData (studentRollNo,studentEmail,studentPassword,studentName,studentSection,studentBatch,studentDept,isHigherStudies,isPlaced) VALUES ('{row['studentRollNo']}','{row['studentEmail']}','{row['studentPassword']}','{row['studentName']}','{row['studentSection']}','{row['studentBatch']}','{row['studentDept']}','{row['isHigherStudies']}','{row['isPlaced']}');\n")

f.close()

```

```python
placement_df = pd.read_csv('placementData.csv')
placement_df
```

```python
placement_df.isnull().sum()

# Print null values
placement_df[placement_df.isnull().any(axis=1)]
```

```python
placement_df['ctc'].fillna(-1, inplace=True)
placement_df['jobRole'].fillna('-', inplace=True)
```

```python
# studentId,companyID,ctc,jobRole,placementDate,isIntern,isPPO,isOnCampus,isGirlsDrive

f = open("insertPlacementData.sql", "w")

for index, row in placement_df.iterrows():
    f.write(f"INSERT INTO placementData (studentId,companyID,ctc,jobRole,placementDate,isIntern,isPPO,isOnCampus,isGirlsDrive) VALUES ({row['studentId']},{row['companyID']},{row['ctc']},'{row['jobRole']}','{row['placementDate']}','{row['isIntern']}','{row['isPPO']}','{row['isOnCampus']}','{row['isGirlsDrive']}');\n")

f.close()
```
